```python
def MAE(N, gt_i_j, pred_i_j):
  MAE = sum(math.abs(gt_i_j - pred_i_j) / N)
```

```python
path = '/content/drive/MyDrive/train_set1/train_dataset_1'
goods = {}
images = []
input_shape = (416, 416)
```

```python
for filename in sorted(os.listdir(path)):
  if filename not in goods:
    for f in filename:
      img = cv2.imread(os.path.join(path + '/' + filename + '/' + f + '.jpg'))
      img = np.array(img)
      images.append(img)
    goods[filename] = images
    images = []
  else: continue
```

```python
import tarfile
zipfile = '/content/data/PublicTestSet (1).tar'
tar = tarfile.open(zipfile, "r:")
tar.extractall('/content/data/')
tar.close()
```

```python
import json
with open("/content/data/PublicTestSet/shelves_info.json", "r") as read_file:
    data = json.load(read_file)
for key in data:
    filename = key
    image = Image.open('/content/data/PublicTestSet/shelves/' + filename)
    img1 = ImageDraw.Draw(image)
    for j in range(len(data[key]['bboxes'])):
        lsd = data[key]['bboxes'][j]
        print(lsd)
        minX = lsd[0]
        minY = lsd[1]
        maxX = lsd[2]
        maxY = lsd[3]
        print(minX)
        print(minY)
        print(maxX)
        print(maxY)
        for i in range(5):
            img1.rectangle([minX+i,minY+i,maxX-i,maxY-i], outline = 'green')
    image = np.array(image)
    im_bgr = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    imgSaveDir = path.join('/content/data_marked', key)
    cv2.imwrite(imgSaveDir, im_bgr)
```
