```python
zipfile = '/content/drive/MyDrive/PublicTestSet.tar'
tar = tarfile.open(zipfile, "r:")
```

```python
tar.extractall('/content/data')
tar.close()
```

```python
import json
with open("/content/data/PublicTestSet/shelves_info.json", "r") as read_file:
    data = json.load(read_file)
for key in data:
    filename = key
    image = Image.open('/content/data/PublicTestSet/shelves/' + filename)
    img1 = ImageDraw.Draw(image)
    for j in range(len(data[key]['bboxes'])):
        lsd = data[key]['bboxes'][j]
        print(lsd)
        minX = lsd[0]
        minY = lsd[1]
        maxX = lsd[2]
        maxY = lsd[3]
        print(minX)
        print(minY)
        print(maxX)
        print(maxY)
        for i in range(5):
            img1.rectangle([minX+i,minY+i,maxX-i,maxY-i], outline = 'green')
    image = np.array(image)
    im_bgr = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    imgSaveDir = path.join('/content/data_marked', key)
    cv2.imwrite(imgSaveDir, im_bgr)
```
